Criao da tabela (Carga)

```
df.createOrReplaceTempView("base")
```

```
%sql

select * from base
```

1 - Qual o nmero de kills(inimigos derrubados) do jogador que teve o maior nmero de MVP(prmio de melhor jogador do round)?

```
%sql

select name, kill_count, mvp from base
where mvp = (select max(mvp) from base)
  
```

2 - Qual o nmero de kills por headhost(inimigos derrubados com tiro na cabea) teve o jogador com o maior nmero de kills(inimigos derrubados)?

```
%sql

select name, headshot_count, kill_count from base
where kill_count = (select max(kill_count) from base)
```

3 - Quantos MVP(prmio de melhor jogador do round) teve o jogador com o maior ADR(mdia de dano por round)?

```
%sql

select name, mvp, adr from base
where adr = (select max(adr) from base)
```

4 - Qual o nmero de deaths(mortes) do jogador com o maior nmero de first kills(derrubou o primeiro inimigo do round)?

```
%sql

select name, death_count, first_kill_count from base
where first_kill_count = (select max(first_kill_count) from base)
```
